```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
```

## Hyperplanes

Toy classification problem in two dimensions

```python
n_samples = 200
n_features = 2
n_classes = 2
n_clusters = 1
class_sep = 1.0
X, y = make_classification(n_samples=n_samples, 
                          n_features=n_features, 
                          n_classes=n_classes, 
                          n_clusters_per_class=n_clusters, 
                          class_sep=class_sep,
                          n_redundant=0, n_repeated=0, n_informative=2, 
                          random_state=16)

# Create the figure and set size
fig = plt.figure(figsize=(5, 5))

# The 111 argument specifies that the subplot grid has 1 row, 
#    1 column, and this subplot is the first & only subplot in the grid.
# The aspect='equal' argument sets the aspect ratio of the subplot to be 
#    equal, meaning that the length along the x-axis is equal to the length 
#    along the y-axis.
ax = fig.add_subplot(111, aspect='equal')

ax.scatter(X[:,0], X[:,1], c=y, alpha=0.6, s=15, cmap='cividis')
ax.arrow(2, -2, -4, 4, width=0.02, head_width=0.0, 
         length_includes_head=True, alpha=1, linestyle='-', color='black')
```

```python
w = np.array([0.1, 0.5])

# setting up the grid
delta = 0.01
xmin=-1
xmax=1
ymin=-1
ymax=1

xs = np.arange(xmin, xmax, delta)
ys = np.arange(ymin, ymax, delta)
x_grid, y_grid = np.meshgrid(xs, ys)

# the value produced by the hyperplane at each point of the grid
Z = w[0] * x_grid + w[1] * y_grid

im = plt.imshow(Z, interpolation='bilinear', origin='lower',
                cmap=plt.cm.gray, extent=(xmin,xmax,ymin,ymax) )

C = plt.contour(x_grid, y_grid, Z, 1, origin='lower',
                colors = 'black')

plt.axhline(y=0, xmin = -1, xmax=1, color='black', alpha=0.5)
plt.axvline(x=0, ymin = -1, ymax=1, color='black', alpha=0.5)

plt.clabel(C, inline=1, fmt='%1.1f')
plt.arrow(0,0,w[0],w[1], width=0.001,head_width=0.05, length_includes_head=True, alpha=1,linestyle='-',color='k');
plt.text(w[0], w[1], r"$\mathbf{w}$");
plt.text(0.2, 0, r"$\mathbf{w}^\top \mathbf{x} = 0$");
```

The sign of the dot product w^Tx can tell us what side on the hyperplane a point x is at

```python
w = np.array([0.1, 0.5])

# setting up the grid
delta = 0.01
xmin=-1;xmax=1;ymin=-1;ymax=1
xs = np.arange(xmin, xmax, delta)
ys = np.arange(ymin, ymax, delta)
x_grid, y_grid = np.meshgrid(xs, ys)

# the value produced by the hyperplane at each point of the grid
Z = w[0] * x_grid + w[1] * y_grid

im = plt.imshow(Z, interpolation='bilinear', origin='lower',
                cmap=plt.cm.gray, extent=(xmin,xmax,ymin,ymax) )

C = plt.contour(x_grid, y_grid, Z, 1, origin='lower',
                colors = 'black')

plt.clabel(C, inline=1, fmt='%1.1f')
plt.arrow(0,0,w[0],w[1], width=0.001,head_width=0.05, length_includes_head=True, alpha=1,linestyle='-',color='k');
plt.text(w[0], w[1], r"$\mathbf{w}$");
plt.text(0.2, 0, r"$\mathbf{w}^\top \mathbf{x} = 0$");

# These lines of code add text annotations to the plot to visually represent the decision boundary of the 
#    linear SVM classifier based on the calculated weights w
plt.text(-0.5, (0.1-w[0]*(-0.5))/w[1], r"$\mathbf{w}^\top\mathbf{x} > 0$");
plt.text(-0.5, (-0.15-w[0]*(-0.5))/w[1], r"$\mathbf{w}^\top\mathbf{x} < 0$"); 
```

## Measuring distance from the hyperplane

```python
w = np.array([0.1, 0.5])

# setting up the grid
delta = 0.01
xmin=-1;xmax=1;ymin=-1;ymax=1
xs = np.arange(xmin, xmax, delta)
ys = np.arange(ymin, ymax, delta)
x_grid, y_grid = np.meshgrid(xs, ys)

Z = w[0] * x_grid + w[1] * y_grid

im = plt.imshow(Z, interpolation='bilinear', origin='lower',
                cmap=plt.cm.gray, extent=(xmin, xmax, ymin, ymax))

C = plt.contour(x_grid, y_grid, Z, 5, 
                origin='lower',
                linewidths=(1, 1, 1, 2, 1, 1, 1),
                colors = 'black')

plt.clabel(C, inline=1, fmt='%1.1f')
 
plt.arrow(0,0,w[0],w[1], width=0.001,head_width=0.05, length_includes_head=True, alpha=1,linestyle='-',color='k');
plt.text(w[0], w[1], r"$\mathbf{w}$");
plt.text(-0.5, (0.1-w[0]*(-0.5))/w[1], r"$\mathbf{w}^\top\mathbf{x} > 0$");
plt.text(-0.5, (-0.15-w[0]*(-0.5))/w[1], r"$\mathbf{w}^\top\mathbf{x} < 0$");
```
